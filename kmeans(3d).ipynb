```
import io
import pandas as pd
df2 = pd.read_csv(io.BytesIO(uploaded['dataset_3d.csv']))
data = df2.values

x = data[:, 0]
y = data[:, 1]
z=data[:,1]
print(z)

def distance(a, b):
    return np.linalg.norm(a - b)

def normalize(x):
    num = x - np.min(x)
    den = np.max(x) - np.min(x)
    return (num / den)

def closest_mean(x, means):
    k = len(means)
    d = float('inf')
    idx = -1
    for i in range(k):
        temp = distance(means[i], x)
        if temp < d:
            d = temp
            idx = i
    return idx

import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d

fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')



ax.scatter(x, y, z, c='b', marker='o', label='blue')


ax.set_xlabel('x axis')
ax.set_ylabel('y axis')
ax.set_zlabel('z axis')
plt.title("3D Scatter Plot Example")
plt.legend()
plt.tight_layout()
plt.show()
```

```
x = normalize(x)
y = normalize(y)
z=normalize(z)


fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')



ax.scatter(x, y, z, c='b', marker='o', label='blue')


ax.set_xlabel('normalized x axis')
ax.set_ylabel('normalized y axis')
ax.set_zlabel('normalized z axis')
plt.title("3D normalized Scatter Plot Example")
plt.legend()
plt.tight_layout()
plt.show()
```

```
l = len(x) 
labels = [0] * l
for i in range(l):
    x1 = x[i]
    y1 = y[i]
    z1=z[i]
    hm = [x1, y1,z1]
    hm = np.array(hm)
    label = closest_mean(hm, means)
    labels[i] = label

labels = np.array(labels)
a, b,c = x[labels == 0], y[labels == 0],z[labels==0]
d, e,f = x[labels == 1], y[labels == 1],z[labels==1]
g, h,i = x[labels == 2], y[labels == 2],z[labels==2]
fig = plt.figure()
ax = fig.add_subplot(111, projection = '3d')


ax.scatter(a, b, c, c='r', marker='.', label='red')
ax.scatter(d, e, f, c='g', marker='.', label='green')
ax.scatter(g, h, i, c='y', marker='.', label='yellow')


ax.scatter(means[0][0], means[0][1], means[0][2], c='r', marker='*', label='red')

ax.scatter(means[1][0], means[1][1], means[1][2], c='g', marker='*', label='green')

ax.scatter(means[2][0], means[2][1], means[2][2], c='y', marker='*', label='yellow')
ax.set_xlabel('normalized x axis')
ax.set_ylabel('normalized y axis')
ax.set_zlabel('normalized z axis')
plt.title("3D normalized Scatter Plot Example with randam 3 means")
plt.legend()
plt.tight_layout()
plt.show()
```

```
import io
import pandas as pd
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
import numpy as np

df2 = pd.read_csv(io.BytesIO(uploaded['dataset.csv']))
data = df2.values

print(data)

def normalize(data):
     return (data - data.min(axis=0))/(data.max(axis=0)-data.min(axis=0))

data = normalize(data)

plt.plot(data[:,0],data[:,1],'.')


means = np.random.random((2,2))


plt.plot(data[:,0],data[:,1],'.')
plt.plot(means[0,0],means[0,1],'ro')
plt.plot(means[1,0],means[1,1],'go')
plt.show()

for i in range(5):
        labels = cdist(data,means).argmin(axis=1)
        for k in range (2):
            dataset = data[labels == k]
            plt.plot(dataset[:,0],dataset[:,1],'.')
            means[k] = dataset.mean(axis=0)
            plt.plot(means[k][0],means[k][1],'X')
        plt.show()
```

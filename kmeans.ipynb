```
import io
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df2 = pd.read_csv(io.BytesIO(uploaded['dataset_3.csv']))
data = df2.values

A = (data[:, 0])
B = (data[:, 1])

plt.plot(A,B,'.')

plt.show()
```

```
def distance(a, b):
      return np.linalg.norm(a - b)
```

```
def normalize(x):
    num = x - np.min(x)
    den = np.max(x) - np.min(x)
    return (num / den)
```

```
def closest_mean(x, means):
    k = len(means)
    d = float('inf')
    idx = -1
    for i in range(k):
        temp = distance(means[i], x)
        if temp < d:
            d = temp
            idx = i
    return idx
```

```
plt.plot(A, B, '.')
plt.title('B and A')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```
A = normalize(A)
B = normalize(B)
plt.plot(A, B, '.')
plt.title('Normalized B and A')
plt.xlabel('A')
plt.ylabel('B')
plt.show()
```

```
K = 3
means = []
for i in range(K):
    x = np.random.rand()
    y = np.random.rand()
    mean = [x, y]
    means.append(mean)

means = np.array(means)

plt.plot(A, B, '.')
plt.plot(means[0][0], means[0][1], 'ro')
plt.plot(means[1][0], means[1][1], 'go')
plt.plot(means[2][0], means[2][1], 'yo')
plt.title('Centroids')
plt.xlabel('A') 
plt.ylabel('B')
plt.show()
```

```
l = len(B)
labels = [0] * l
for i in range(l):
    A1 = A[i]
    B1= B[i]
    x = [A1, B1]
    x = np.array(x)
    label = closest_mean(x, means)
    labels[i] = label

labels = np.array(labels)
a, b = A[labels == 0], B[labels == 0]
c, d = A[labels == 1], B[labels == 1]
e, f = A[labels == 2], B[labels == 2]

plt.plot(a, b, 'r.')
plt.plot(c, d, 'g.')
plt.plot(e, f, '.')
plt.plot(means[0][0], means[0][1], 'ro', markersize=15)
plt.plot(means[1][0], means[1][1], 'go', markersize=15)
plt.plot(means[2][0], means[2][1], 'bo', markersize=15)
plt.plot()
```

```
means[0][0], means[0][1]= a.mean(), b.mean()
means[1][0], means[1][1] = c.mean(), d.mean()
means[2][0], means[2][1] = e.mean(), f.mean()
plt.plot(a, b, 'r.')
plt.plot(means[0][0], means[0][1], 'ro')

plt.plot(c, d, 'g.')
plt.plot(means[1][0], means[1][1], 'go')

plt.plot(e, f, 'b.')
plt.plot(means[2][0], means[2][1], 'bo')


plt.show()
```
